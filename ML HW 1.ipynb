```python
# 1 
import matplotlib.pyplot as plt
import numpy as np

years = np.linspace(1930,2010,9)
gdp = [[1.015], [1.33], [2.29], [3.26], [4.951], [6.759], [9.366], [13.131], [15.599]]

plt.figure()
plt.plot(years,gdp,'r')
plt.xlabel("Year")
plt.ylabel("Trillions of $")
plt.title("Real US GDP (in Trillions)")

```

2. Based on the graph, the shape of the curve appears to be a quadratic polynomial function. We will use Python's polyfit function to determine an approximation for the equation of the function. Python's polyfit function uses the least squares estimation method to determine a best fit approximation for the data. Since we assume that the data in quadratic, the third input into the polyfit function is 2. 



```python
best = np.polyfit(years,gdp,2)
print(best)
```

The output of the polyfit function is shown above. This means that the approximation of the function is 2.206*10^-3*x^2 - 8.506x + 8200


```python
#3
y = []
x = years
for i in range (9):
    ytemp = best[0]*(x[i]**2)+best[1]*x[i]+best[2]
    y.append(ytemp)
plt.plot(years,gdp,'r')    
plt.hold
plt.plot(years,y)
plt.xlabel("Year")
plt.ylabel("Trillions of $")
plt.title("Real US GDP (in Trillions)")
```

The red line the given GDP data. The blue line is the function y=2.206*10^-3*x^2 - 8.506x + 8200 that was mathematically computed with Python. As we can see, this function provides an excellent approximation of the given data.
